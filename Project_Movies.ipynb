```python
df_ml1=pd.merge(df_ratings,df_users,on='UserID')
df_ml=pd.merge(df_movies,df_ml1,on='MovieID')
```

```python
df_ml
```

```python
from sklearn.utils import shuffle
df_ml_shuffle = shuffle(df_ml)
df_ml_500=df_ml_shuffle.head(500)
```

```python
df_ml_500
```

```python
import pandas as pd
import statsmodels.formula.api as sm
```

```python
model=sm.ols(formula='Rating~MovieID+Age+Occupation',data=df_ml_500)
model_fit=model.fit()
model_fit.summary()
```

```python
model_fit.predict()
```

```python
import pandas.plotting as plt2
plt2.scatter_matrix(df_ml_500[['Rating','MovieID','Age','Occupation']],figsize=(8,8),color="green",diagonal="hist")
```

```python
import seaborn as sns
cor_coff=df_ml_500[['Rating','MovieID','Age','Occupation']].corr()
sns.heatmap(cor_coff,annot=True,cmap="winter_r")
```
